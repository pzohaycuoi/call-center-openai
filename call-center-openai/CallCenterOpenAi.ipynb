```python
from os import getenv
from dotenv import load_dotenv
import chatgpt_api
import microphone_text
```

```python
load_dotenv()
text = microphone_text.recognize_from_microphone(getenv('SPEECH_KEY'),
                                                 getenv('SPEECH_REGION'))
result = chatgpt_api.chat_completion(getenv('ORG_ID'),
                                     getenv('OPENAI_API_KEY'), text)
print(f"ChatGPT: {result['choices'][0]['message']['content']}")
print(result)
```
